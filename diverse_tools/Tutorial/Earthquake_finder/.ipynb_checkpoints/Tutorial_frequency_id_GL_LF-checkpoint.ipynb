```python
def silhouette_score_estimation(df_kmeans_input):
    
    '''    
    
    Compute the mean Silhouette Coefficient.

    df_kmeans_input: dataframe with the Welch high frequency power spectral density (PSD) estimates

    '''
    X_train = TimeSeriesResampler(sz=len(df_kmeans_input['data_FFT_HF'].values[0])).fit_transform(df_kmeans_input['data_FFT_HF'].values)

    sscore = silhouette_score(X_train, df_kmeans_input['k-means'], metric="euclidean")

    return round(sscore,2)


def Elbow_KMeans_estimation(df_kmeans_input):

    '''
    Compute the optimal K value in a k-means clustering algorithm.

    df_kmeans_input: dataframe with the Welch high frequency power spectral density (PSD) estimates
    '''
    
    # ------------------------------------
    # Adjusting arrays to KMeans procedure
    X_train = TimeSeriesResampler(sz=len(df_kmeans_input['data_FFT_HF'].values[0])).fit_transform(df_kmeans_input['data_FFT_HF'].values)
    
    # --------------------------------
    # Elbow KMeans inertias estimation

    elbow_data = []
    for n_clusters in tqdm(range(1,10,1),total=len(range(1,10,1)),desc='Elbow Method: ',position=0,leave=False):
                
        km = TimeSeriesKMeans(n_clusters=n_clusters, verbose=False, random_state=0,n_jobs=20)
        y_pred = km.fit_predict(X_train)
        elbow_data.append((n_clusters, km.inertia_))
            
    k_range = []
    inertias = []
    for elb in elbow_data:
        k_range.append(elb[0])
        inertias.append(elb[1])
    
    # ----------------------------   
    # Elbow KMeans Knee estimation
    kn = KneeLocator(k_range, inertias,S=2, curve='convex', direction='decreasing')
    elbow_point = kn.knee+1

    return elbow_point,inertias,k_range

def Elbow_KMeans_plot(elbow_point,inertias,k_range):

    '''
    elbow_point: Estimated number of clusters.
    k_range: Number of Clusters (k).
    inertias: Inertia is the sum of squared distances of each data point to its closest cluster center. 
              It’s essentially the total squared error of the clustering.
              A lower inertia implies that the data points are more closely grouped around their respective cluster centers. 
    '''
    
    # ------------------------------------
    # Elbow KMeans plot 
                
    fig_elb, ax_elb = plt.subplots(figsize=(10,5))
                
    ax_elb.annotate(str(elbow_point), xy=(elbow_point,inertias[k_range.index(elbow_point)]), xytext=(elbow_point+(elbow_point*0.5),inertias[k_range.index(elbow_point)]+(inertias[k_range.index(elbow_point)]*0.5)),arrowprops=dict(arrowstyle='<-',linewidth=0.25,color='black'))
    ax_elb.scatter(elbow_point,inertias[k_range.index(elbow_point)],color='k', marker='X',s=100)
    ax_elb.plot(k_range,inertias,color='grey', marker='o', linestyle='dashed',linewidth=2, markersize=5,zorder=-1)
    ax_elb.text(0.85, 0.9,'LFI',fontweight='bold', fontsize=25, ha='right',va='center',transform=ax_elb.transAxes)

    # Set ticks and labels on both sides of the y-axis
    ax_elb.set_xlim(0,10)
    ax_elb.set_ylabel('Inertias')
    ax_elb.set_xlabel('Values of K')
    fig_elb.savefig(FOLDER_OUTPUT+'FIGURAS/PSD/FINAL/LF_'+dep+'_Elbow_plot.png',pad_inches=0.02,dpi=200)           


def KMeans_clustering_and_plot(df_kmeans_input,n_clu):

    '''
    df_kmeans_input: dataframe with the Welch high frequency power spectral density (PSD) estimates
    '''
    # ------------------------------------
    # Adjusting arrays to KMeans procedure
    X_train = TimeSeriesResampler(sz=len(df_kmeans_input['data_FFT_HF'].values[0])).fit_transform(df_kmeans_input['data_FFT_HF'].values)
        
    # ----------------------------------------------------------------
    # Euclidean k-means with the estimated number of clusters and plot
               
    km = TimeSeriesKMeans(n_clusters=n_clu, verbose=False, random_state=0,n_jobs=20)
    y_pred = km.fit_predict(X_train)
                
    df_kmeans_input['k-means'] = y_pred

    # ---------------------------------------
    # Compute the mean Silhouette Coefficient 

    sscore = silhouette_score_estimation(df_kmeans_input)
    
    ################################
    ##### CREATING THE FIGURE ######
    ################################
    
    # ----------
    # Parameters
    
    lw = 1
    alpha = 0.5
    
    # Locator: X & Y
    majorX = MultipleLocator(5)
    majorY = MultipleLocator(20)
                
    minorX = MultipleLocator(1)
    minorY = MultipleLocator(10)
    
    # ----------
    
    fig = plt.figure(figsize=(20,10))
    gs = gridspec.GridSpec(3, n_clu)
                
    ax_psd_HF = fig.add_subplot(gs[:2, :])
                   
    # Plot PSD computed via both methods
    
    # Define unique colors for each category in 'LFI'
    unique_labels = df_kmeans_input['LFI'].unique()
    cmap = cm.get_cmap('turbo', len(unique_labels))  # Define a discrete colormap with a fixed number of colors
    color_map = {label: cmap(i) for i, label in enumerate(unique_labels)}
    
    for i in df_kmeans_input.iterrows():
        year = i[1]['LFI']
        color = color_map[year]
        ax_psd_HF.plot(i[1]['freq_FFT_HF'], i[1]['data_FFT_HF'], color=color, lw=lw, alpha=alpha)
                
    ax_psd_HF.text(0.9, 0.9,'SS: '+str(sscore)+'\n'+dep+' [n:'+str(kmeans_df_dep.shape[0])+']',fontweight='bold', fontsize=12, ha='center',va='center',transform=ax_psd_HF.transAxes)
               
    ax_psd_HF.set_ylabel('Spectral level (dB)')
                
    # Set limits for x-axes
    ax_psd_HF.set_xlim(1, 25)
    ax_psd_HF.set_ylim(-140,-20)
                
    # Set ticks and labels on both sides of the y-axis
    ax_psd_HF.yaxis.set_tick_params(labelright=True,left=True, right=True)  # Show labels on the right side
    ax_psd_HF.spines['top'].set_linewidth(2)
    ax_psd_HF.spines['right'].set_linewidth(2)
    ax_psd_HF.spines['bottom'].set_linewidth(2)
    ax_psd_HF.spines['left'].set_linewidth(2)
    ax_psd_HF.xaxis.set_tick_params(width=2)
    ax_psd_HF.yaxis.set_tick_params(width=2)
                
    ax_psd_HF.xaxis.set_major_locator(majorX)
    ax_psd_HF.xaxis.set_minor_locator(minorX)
    ax_psd_HF.yaxis.set_major_locator(majorY)
    ax_psd_HF.yaxis.set_minor_locator(minorY)
    
    # =======================================================================================================================================
                
    # Get unique k-means
    unique_k_means = sorted(df_kmeans_input['k-means'].unique())
                
    for k in unique_k_means:
        ax = fig.add_subplot(gs[2, k])
                
        df_k = df_kmeans_input[df_kmeans_input['k-means'] == k]
                
        # Plot PSD computed via both methods
        for i in df_k.iterrows():
            year = i[1]['LFI']
            color = color_map[year]
            ax.plot(i[1]['freq_FFT_HF'], i[1]['data_FFT_HF'], color=color, lw=lw, alpha=alpha)
        ax.plot(df_k['freq_FFT_HF'].mean(),df_k['data_FFT_HF'].mean(), color='k', lw=2,ls='--')
        ax.text(0.5, 0.8,'Cluster %d' % (k + 1)+'\n(n='+str(df_k.shape[0])+')',fontweight='bold', fontsize=12, ha='center',transform=ax.transAxes)
                
        ax.set_xlabel('Frequency (Hz)')
        ax.set_ylim(-140,-20)
        ax.set_xlim(1, 25)
                
        ax.spines['top'].set_linewidth(2)
        ax.spines['right'].set_linewidth(2)
        ax.spines['bottom'].set_linewidth(2)
        ax.spines['left'].set_linewidth(2)
        ax.xaxis.set_tick_params(width=2)
        ax.yaxis.set_tick_params(width=2)  
                
        ax.xaxis.set_major_locator(majorX)
        ax.xaxis.set_minor_locator(minorX)
        ax.yaxis.set_major_locator(majorY)
        ax.yaxis.set_minor_locator(minorY)
        ax.yaxis.set_tick_params(labelright=False,labelleft=False,left=True, right=True)  # Show labels on the right side
                
        if k == unique_k_means[0]:
            ax.yaxis.set_tick_params(labelright=False,labelleft=True,left=True, right=True)  # Show labels on the right side
            ax.set_ylabel('Spectral level (dB)')
                
        if k == unique_k_means[-1]:
            ax.yaxis.set_tick_params(labelright=True,labelleft=False,left=True, right=True)  # Show labels on the right side
         
    # Create a color bar
    # Normalize labels for color bar
    norm = mcolors.BoundaryNorm(boundaries=np.arange(len(unique_labels)+1)-0.5, ncolors=len(unique_labels))
    sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
    
    # Create the colorbar and set labels
    cbar = fig.colorbar(sm, ax=ax_psd_HF, orientation='vertical', ticks=np.arange(len(unique_labels)))
    cbar.set_label('LFI code')
    cbar.set_ticks(np.arange(len(unique_labels)))  # Ensure labels are at discrete points
    cbar.set_ticklabels(unique_labels)  # Set tick labels to match LFI codes
    fig.savefig(FOLDER_OUTPUT+'FIGURAS/PSD/FINAL/LF_'+dep+'_kmeans_plot.png',pad_inches=0.02,dpi=200)
    
    return df_kmeans_input
```
